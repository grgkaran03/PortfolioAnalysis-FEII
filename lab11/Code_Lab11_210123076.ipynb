```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

### Q1. Vasicek Model

```python
def VasicekYield(t,T,beta,mu,sigma,r):
    b = beta*mu
    a = beta
    B = (1 - np.exp(-a*(T-t)))/a
    A = ((B - T + t)*(a*b-0.5*sigma*sigma)/(a*a)) - ((sigma*sigma)*(B*B)/(4*a))

    p = np.exp(A - B*r)
    return- np.log(p)/(T-t)
```

```python
params = [[5.9, 0.2, 0.3, 0.1], [3.9, 0.1, 0.3, 0.2], [0.1, 0.4, 0.11, 0.1]]

T_ = np.linspace(0.1,10,10,endpoint=False)

i = 0
for param in params:
    beta, mu, sigma, r = param  
    y= []
    for T in T_:
        y.append(VasicekYield(0,T,beta,mu,sigma,r))
    plt.plot(T_, y, marker='o')
    plt.xlabel('Maturity (T)')
    plt.ylabel('Yield')
    plt.title('Term structure for parameter set - {}'.format(i + 1))
    plt.show()

    i+=1
        
```

### Q2. CIR Model

```python
def CirYield(t, T,beta, mu, sigma, r):
    gamma = math.sqrt(beta*beta + 2*sigma*sigma)
    B = 2 *(math.exp(gamma * (T - t)) - 1)  / ( 2*gamma + (gamma + beta) * (math.exp(gamma * (T - t)) - 1))
    A = math.pow( ( 2*gamma*math.exp(0.5*(beta + gamma)*(T - t)) ) / (2*gamma + (gamma + beta)*(math.exp(gamma*(T - t)) - 1)), 2*beta*mu / (sigma*sigma) ) 
    P = A * math.exp(-B * r)
    y = -math.log(P) / (T - t)

    return y
```

```python
params = [[0.02, 0.7, 0.02, 0.1], [0.7, 0.1, 0.3, 0.2], [0.06, 0.09, 0.5, 0.02]]

T_ = np.linspace(0.1,10,10,endpoint=False)

i = 0
for param in params:
    beta, mu, sigma, r = param  
    y= []
    for T in T_:
        y.append(CirYield(0,T,beta,mu,sigma,r))
    plt.plot(T_, y, marker='o')
    plt.xlabel('Maturity (T)')
    plt.ylabel('Yield')
    plt.title('Term structure for parameter set - {}'.format(i + 1))
    plt.show()

    i+=1
```
